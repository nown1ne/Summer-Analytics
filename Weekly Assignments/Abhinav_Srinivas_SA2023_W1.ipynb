3. Which director has directed the most number of movies?

```python
counts = movies['Director'].value_counts()
most_common_director = counts.idxmax()
num_movies = counts.max()

print("Director with the most number of movies:", most_common_director)
print("Number of movies directed:", num_movies)
```

4. Name the Movie which has the highest Runtime ?

```python
movies['Runtime'] = pd.to_timedelta(movies['Runtime'])

runtime_movie = movies.loc[movies['Runtime'].idxmax(), 'Series_Title']
runtime = movies['Runtime'].max()

runtime_str = str(runtime)

print("Movie with the highest runtime:", runtime_movie)
print("Runtime:", runtime_str[:-3])
```

5. How many movies are released after year 2000 and have IMDB rating > 8.5 ?





```python
movies['Released_Year'] = pd.to_numeric(movies['Released_Year'], errors='coerce')

filtered_movies = movies[(movies['Released_Year'] > 2000) & (movies['IMDB_Rating'] > 8.5)]

num_movies = len(filtered_movies)

print("Number of movies released after 2000 with IMDb rating > 8.5:", num_movies)
```

6. Make a countplot of Movies directed by all directors.

```python
import plotly.graph_objects as go
director_counts = movies['Director'].value_counts()

bar_plot = go.Bar(
    x=director_counts.index,
    y=director_counts.values
)

layout = go.Layout(
    title='Movies Directed by Directors',
    xaxis=dict(title='Director'),
    yaxis=dict(title='Number of Movies')
)

fig = go.Figure(data=[bar_plot], layout=layout)
fig.show()
```

```python
plt.figure(figsize=(10, 6))  

top_directors = movies['Director'].value_counts().nlargest(5).index

sns.countplot(data=movies[movies['Director'].isin(top_directors)], x='Director')

plt.xticks(rotation=90)  
plt.xlabel('Director')  
plt.ylabel('Count')  
plt.title('Number of Movies Directed by Top 5 Directors')  

plt.show()
```

7. Make a plot to analyze distribution of IMDB rating?

```python
plt.figure(figsize=(8, 6))  

sns.histplot(data=movies, x='IMDB_Rating', bins=10, kde=True)

plt.xlabel('IMDB Rating') 
plt.ylabel('Count')  
plt.title('Distribution of IMDB Ratings')  

plt.show()  
```

8. What type of Genre has the highest IMDB rating?


```python
movies['Split_Genres'] = movies['Genre'].str.split(',')

exploded_genres = movies.explode('Split_Genres')
genre_avg_rating = exploded_genres.groupby('Split_Genres')['IMDB_Rating'].mean().sort_values(ascending=False)

top_genre = genre_avg_rating.idxmax()

plt.figure(figsize=(10, 6))
sns.barplot(x=genre_avg_rating.values, y=genre_avg_rating.index, palette='viridis')
plt.xlabel('Average IMDb Rating')
plt.ylabel('Genre')
plt.title('Average IMDb Rating by Genre')
plt.axvline(x=genre_avg_rating.loc[top_genre], color='red', linestyle='--', label='Highest Rating Genre')
plt.legend()
plt.show()

print("Genre with the highest IMDb rating:", top_genre)
```

9. Name the top 5 movies that have earned highest Meta score. 

```python
top_movies = movies.nlargest(5, 'Meta_score')

plt.figure(figsize=(10, 6))
sns.barplot(x='Series_Title', y='Meta_score', data=top_movies, palette='viridis')
plt.xlabel('Movie')
plt.ylabel('Meta Score')
plt.title('Top 5 Movies with Highest Meta Score')
plt.show()
```

10. Of which Genre movie has the highest Runtime?

```python
movies['Split_Genres'] = movies['Genre'].str.split(',')
genre_runtime = movies.explode('Split_Genres')

genre_runtime['Runtime_minutes'] = genre_runtime['Runtime'].dt.total_seconds() / 60

top_genre = genre_runtime.groupby('Split_Genres')['Runtime_minutes'].max().idxmax()

movie_highest_runtime = genre_runtime[genre_runtime['Split_Genres'] == top_genre].iloc[0]
plt.figure(figsize=(10, 6))
sns.barplot(x='Runtime_minutes', y='Split_Genres', data=genre_runtime, palette='viridis')
plt.xlabel('Runtime (minutes)')
plt.ylabel('Genre')
plt.title('Runtime Distribution by Genre')
plt.show()
```
